## Part 3: Make a combined visualization

We want to make a graph that shows months in which Chris has most data and compare his data frequencey for these months with Mosalam.

```python
# initialize both dataframes
df_Chris = df_partner_2
df_Mosalam = df_partner
```

```python
# First, we need to convert the time to year-month format
Chris = df_Chris["endTime"].dt.to_period('M')

# Then, we will create a dataframe of Chris frequency in each month in his timestamps
df_Chris_2 = pd.DataFrame(Chris.value_counts())
df_Chris_2.set_index("endTime")
df_Chris_2 = df_Chris_2.rename(columns={'endTime': 'Frequency'})
df_Chris_2
```

```python
# We apply the same stuff that we've done for Chris's data
Mosalam = df_Mosalam["Time"].dt.to_period('M')
df_Mosalam_2 = pd.DataFrame(Mosalam.value_counts())
df_Mosalam_2.set_index("Time")
df_Mosalam_2 = df_Mosalam_2.rename(columns={'Time': 'Frequency'})
df_Mosalam_2
```

```python
# Now, we'll merge both dataframes
df_merged = df_Chris_2.merge(df_Mosalam_2, left_index=True, right_index=True)
df_merged
```

```python
# Rename the columns to make more sense and sort the values ascendingly by the date
df_merged.rename(columns={'Frequency_x': 'Frequency_Chris', 'Frequency_y': 'Frequency_Mosalam'}, inplace = True)
df_merged.sort_index(inplace=True)

# Convert the index from time to string so that we can use index values for graphing
df_merged.index = df_merged.index.strftime('%Y-%m')
df_merged
```

```python
plt.scatter(df_merged.index, df_merged['Frequency_Chris'], label='Chris values', color = "green")
plt.scatter(df_merged.index, df_merged['Frequency_Mosalam'], label='Mosalam values')
plt.xticks(df_merged.index, df_merged.index, rotation=90)
plt.xlabel("Dates")
plt.ylabel("Frequency")

plt.show()
```

The graph doesn't show enough details for Mosalam's values because they are much lower than Chris's values.

To fix that, we will use two different y-axis scales for each group values.

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.scatter(df_merged.index, df_merged['Frequency_Chris'], label='Chris values', color = "green")
ax2.scatter(df_merged.index, df_merged['Frequency_Mosalam'], label='Mosalam values', color='red')
ax1.set_xlabel('Dates')
ax1.set_xticks(df_merged.index, df_merged.index, rotation=90)
ax1.set_ylabel('Frequency_Chris', color='green')
ax2.set_ylabel('Frequency_Mosalam', color='red')

plt.show()
```

- From the first graph, we can tell that Chris has much more values per month than Mosalam.

- From the second graph, we can see which months has the highest frequencies of data for each of us. 

- For example, lowest point for Chris's values is at 2023-3, so this represents the month in which he had the least amount data: about 500 values. While for Mosalam, it's at 2002-5 with around 3 values in that month.

### Chris's Part:

Count the number of songs in each of my playlists. How many Songs are in EACH of my playlists?

```python
df_part4 = pd.read_json("./Spotify/MyData/Playlist1.json")
freq = []
playlist_num = []
count = 1
for i in df_part4["playlists"]:
    num_tracks = 0
    for j in i["items"]:
        if 'track' in j:
            num_tracks += 1
    freq.append(num_tracks)
    playlist_num.append(count)
    count += 1
d = {"Playlist" : playlist_num, "Frequency" : freq}
df_freq = pd.DataFrame(d)
df_freq
```

```python
plt.bar(df_freq["Playlist"], df_freq["Frequency"], color="green")
plt.xlabel("Playlist Number")
plt.ylabel("Number of Tracks")
plt.title("Number of Songs in Chris' Spotify Playlists")
```
